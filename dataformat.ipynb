```python
import pandas as pd
import pandas.plotting._matplotlib
```

```python
oly_2004 = pd.read_csv('./data/olympic_2004.csv')
oly_2008 = pd.read_csv('./data/olympic_2008.csv')
oly_2012 = pd.read_csv('./data/olympic_2012.csv')
oly_2016 = pd.read_csv('./data/olympic_2016.csv')
oly_2020 = pd.read_csv('./data/olympic_2020.csv')
```

```python
SHO = pd.read_csv('./data/SHO.csv')
SWIM = pd.read_csv('./data/SWIM.csv')
WLF = pd.read_csv('./data/WLF.csv')
GAR = pd.read_csv('./data/GAR.csv')
ATH = pd.read_csv('./data/ATH.csv')

SHO_2004 = pd.read_csv('./data/sho_2004.csv')
SWIM_2004 = pd.read_csv('./data/swim_2004.csv')
WLF_2004 = pd.read_csv('./data/weight_2004.csv')
GAR_2004 = pd.read_csv('./data/gar_2004.csv')
ATH_2004 = pd.read_csv('./data/ath_2004.csv')
SHO_2008 = pd.read_csv('./data/sho_2008.csv')
SWIM_2008 = pd.read_csv('./data/swim_2008.csv')
WLF_2008 = pd.read_csv('./data/weight_2008.csv')
GAR_2008 = pd.read_csv('./data/gar_2008.csv')
ATH_2008 = pd.read_csv('./data/ath_2008.csv')
SHO_2012 = pd.read_csv('./data/sho_2012.csv')
SWIM_2012 = pd.read_csv('./data/swim_2012.csv')
WLF_2012 = pd.read_csv('./data/weight_2012.csv')
GAR_2012 = pd.read_csv('./data/gar_2012.csv')
ATH_2012 = pd.read_csv('./data/ath_2012.csv')
SHO_2016 = pd.read_csv('./data/sho_2016.csv')
SWIM_2016 = pd.read_csv('./data/swim_2016.csv')
WLF_2016 = pd.read_csv('./data/weight_2016.csv')
GAR_2016 = pd.read_csv('./data/gar_2016.csv')
ATH_2016 = pd.read_csv('./data/ath_2016.csv')
SHO_2020 = pd.read_csv('./data/sho_2020.csv')
SWIM_2020 = pd.read_csv('./data/swim_2020.csv')
WLF_2020 = pd.read_csv('./data/weight_2020.csv')
GAR_2020 = pd.read_csv('./data/gar_2020.csv')
ATH_2020 = pd.read_csv('./data/ath_2020.csv')
```

```python
oly_2004['Year'] = 2004
oly_2008['Year'] = 2008
oly_2012['Year'] = 2012
oly_2016['Year'] = 2016
oly_2020['Year'] = 2020
```

```python
oly_df = pd.concat([oly_2004, oly_2008, oly_2012, oly_2016, oly_2020])
```

```python
oly_df.head(3)
```

```python
import json
with open('./data/world.geojson') as f:
    data = json.load(f)
```

```python
for line in data['features']:
    countryID = line['id']
    if countryID == "DEU":
        countryID = "GER"
    if countryID == "RUC":
        countryID = "ROC"
    if countryID == "PRT":
        countryID = "POR"
    if countryID == "NLD":
        countryID = "NED"
    if countryID == "DEN":
        countryID = "DNK"

    countryDF = oly_df[oly_df['NOC'] == countryID].sort_values(by=['Year'])

    line['medals'] = {
                    "Gold": countryDF[" Gold"].tolist() if len(countryDF) > 0 else [0, 0, 0, 0, 0],
                    "Silver": countryDF[" Silver"].tolist() if len(countryDF) > 0 else [0, 0, 0, 0, 0],
                    "Bronze": countryDF[" Bronze"].tolist() if len(countryDF) > 0 else [0, 0, 0, 0, 0],
                    "Total": countryDF[" Total"].tolist() if len(countryDF) > 0 else [0, 0, 0, 0, 0],
                    "Year": countryDF["Year"].tolist() if len(countryDF) > 0 else [0, 0, 0, 0, 0]
        
                    }
    line['average'] = {
                    "Gold": countryDF[" Gold"].astype(int).sum() /len(countryDF)  if len(countryDF) > 0 else 0,
                    "Silver": countryDF[" Silver"].astype(int).sum() /len(countryDF)  if len(countryDF) > 0 else 0,
                    "Bronze": countryDF[" Bronze"].astype(int).sum() /len(countryDF)  if len(countryDF) > 0 else 0,
                    "Total": countryDF[" Total"].astype(int).sum() /len(countryDF)  if len(countryDF) > 0 else 0
                    }

    SHO_CTRY = SHO[SHO['NOC'] == countryID]
    SWIM_CTRY = SWIM[SWIM['NOC'] == countryID]
    WLF_CTRY = WLF[WLF['NOC'] == countryID]
    GAR_CTRY = GAR[GAR['NOC'] == countryID]
    ATH_CTRY = ATH[ATH['NOC'] == countryID]

    line['fields'] = {
                    "SHO": SHO_CTRY["Total"].tolist()[0] if len(SHO_CTRY) > 0 else 0,
                    "SWIM": SWIM_CTRY["Total"].tolist()[0] if len(SWIM_CTRY) > 0 else 0,
                    "WLF": WLF_CTRY["Total"].tolist()[0] if len(WLF_CTRY) > 0 else 0,
                    "GAR": GAR_CTRY["Total"].tolist()[0] if len(GAR_CTRY) > 0 else 0,
                    "ATH": ATH_CTRY["Total"].tolist()[0] if len(ATH_CTRY) > 0 else 0
    }
    SHO_CTRY_2004 = SHO_2004[SHO_2004['NOC'] == countryID]
    SWIM_CTRY_2004 = SWIM_2004[SWIM_2004['NOC'] == countryID]
    WLF_CTRY_2004 = WLF_2004[WLF_2004['NOC'] == countryID]
    GAR_CTRY_2004 = GAR_2004[GAR_2004['NOC'] == countryID]
    ATH_CTRY_2004 = ATH_2004[ATH_2004['NOC'] == countryID]
    
    SHO_CTRY_2008 = SHO_2008[SHO_2008['NOC'] == countryID]
    SWIM_CTRY_2008 = SWIM_2008[SWIM_2008['NOC'] == countryID]
    WLF_CTRY_2008 = WLF_2008[WLF_2008['NOC'] == countryID]
    GAR_CTRY_2008 = GAR_2008[GAR_2008['NOC'] == countryID]
    ATH_CTRY_2008 = ATH_2008[ATH_2008['NOC'] == countryID]
    
    SHO_CTRY_2012 = SHO_2012[SHO_2012['NOC'] == countryID]
    SWIM_CTRY_2012 = SWIM_2012[SWIM_2012['NOC'] == countryID]
    WLF_CTRY_2012 = WLF_2012[WLF_2012['NOC'] == countryID]
    GAR_CTRY_2012 = GAR_2012[GAR_2012['NOC'] == countryID]
    ATH_CTRY_2012 = ATH_2012[ATH_2012['NOC'] == countryID]
    
    SHO_CTRY_2016 = SHO_2016[SHO_2016['NOC'] == countryID]
    SWIM_CTRY_2016 = SWIM_2016[SWIM_2016['NOC'] == countryID]
    WLF_CTRY_2016 = WLF_2016[WLF_2016['NOC'] == countryID]
    GAR_CTRY_2016 = GAR_2016[GAR_2016['NOC'] == countryID]
    ATH_CTRY_2016 = ATH_2016[ATH_2016['NOC'] == countryID]

    SHO_CTRY_2020 = SHO_2020[SHO_2020['NOC'] == countryID]
    SWIM_CTRY_2020 = SWIM_2020[SWIM_2020['NOC'] == countryID]
    WLF_CTRY_2020 = WLF_2020[WLF_2020['NOC'] == countryID]
    GAR_CTRY_2020 = GAR_2020[GAR_2020['NOC'] == countryID]
    ATH_CTRY_2020 = ATH_2020[ATH_2020['NOC'] == countryID]

    
    line['2004'] = {
                    "SHO": SHO_CTRY_2004["Total"].tolist()[0] if len(SHO_CTRY_2004) > 0 else 0,
                    "SWIM": SWIM_CTRY_2004["Total"].tolist()[0] if len(SWIM_CTRY_2004) > 0 else 0,
                    "WLF": WLF_CTRY_2004["Total"].tolist()[0] if len(WLF_CTRY_2004) > 0 else 0,
                    "GAR": GAR_CTRY_2004["Total"].tolist()[0] if len(GAR_CTRY_2004) > 0 else 0,
                    "ATH": ATH_CTRY_2004["Total"].tolist()[0] if len(ATH_CTRY_2004) > 0 else 0
    }
    
    
    line['2008'] = {
                    "SHO": SHO_CTRY_2008["Total"].tolist()[0] if len(SHO_CTRY_2008) > 0 else 0,
                    "SWIM": SWIM_CTRY_2008["Total"].tolist()[0] if len(SWIM_CTRY_2008) > 0 else 0,
                    "WLF": WLF_CTRY_2008["Total"].tolist()[0] if len(WLF_CTRY_2008) > 0 else 0,
                    "GAR": GAR_CTRY_2008["Total"].tolist()[0] if len(GAR_CTRY_2008) > 0 else 0,
                    "ATH": ATH_CTRY_2008["Total"].tolist()[0] if len(ATH_CTRY_2008) > 0 else 0
    }
    line['2012'] = {
                    "SHO": SHO_CTRY_2012["Total"].tolist()[0] if len(SHO_CTRY_2012) > 0 else 0,
                    "SWIM": SWIM_CTRY_2012["Total"].tolist()[0] if len(SWIM_CTRY_2012) > 0 else 0,
                    "WLF": WLF_CTRY_2012["Total"].tolist()[0] if len(WLF_CTRY_2012) > 0 else 0,
                    "GAR": GAR_CTRY_2012["Total"].tolist()[0] if len(GAR_CTRY_2012) > 0 else 0,
                    "ATH": ATH_CTRY_2012["Total"].tolist()[0] if len(ATH_CTRY_2012) > 0 else 0
    }
    line['2016'] = {
                    "SHO": SHO_CTRY_2016["Total"].tolist()[0] if len(SHO_CTRY_2016) > 0 else 0,
                    "SWIM": SWIM_CTRY_2016["Total"].tolist()[0] if len(SWIM_CTRY_2016) > 0 else 0,
                    "WLF": WLF_CTRY_2016["Total"].tolist()[0] if len(WLF_CTRY_2016) > 0 else 0,
                    "GAR": GAR_CTRY_2016["Total"].tolist()[0] if len(GAR_CTRY_2016) > 0 else 0,
                    "ATH": ATH_CTRY_2016["Total"].tolist()[0] if len(ATH_CTRY_2016) > 0 else 0
    }
    line['2020'] = {
                    "SHO": SHO_CTRY_2020["Total"].tolist()[0] if len(SHO_CTRY_2020) > 0 else 0,
                    "SWIM": SWIM_CTRY_2020["Total"].tolist()[0] if len(SWIM_CTRY_2020) > 0 else 0,
                    "WLF": WLF_CTRY_2020["Total"].tolist()[0] if len(WLF_CTRY_2020) > 0 else 0,
                    "GAR": GAR_CTRY_2020["Total"].tolist()[0] if len(GAR_CTRY_2020) > 0 else 0,
                    "ATH": ATH_CTRY_2020["Total"].tolist()[0] if len(ATH_CTRY_2020) > 0 else 0
    }
    if (len(countryDF) == 0):
        print(countryID)
    
    
```
